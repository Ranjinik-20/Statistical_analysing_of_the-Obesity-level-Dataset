```python
# Statistic and Trends -  Using a dataset called Obesity dataset various plots as been plotted such as Scatterplot, Histogram, boxplot plot and Heatmap using correlation matrix
```

```python
# Importing the required libraries for Plotting

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the dataset using 'pandas' method and Replacing the term from 'Public_Transportation' to 'PB'

df = pd.read_csv("ObesityDataSet.csv")
df.Transportation = df.Transportation.replace('Public_Transportation', 'PB')
df
```

```python
# Plotting the scatterplot for the Relational Graph, for the columns Age and Weight with Obesity level 

def scatterplot_obesity_level(df):
    '''
    Creates subplots for physical activity according to age and weight.
    '''
    obesity_levels = df['Obesity level'].unique()
    num_levels = len(obesity_levels)
    num_cols = (num_levels + 1) // 2  
    fig, axes = plt.subplots(nrows=2, ncols=num_cols, figsize=(5 * num_cols, 10), sharey=True)
    axes = axes.flatten()
    
    # Iterate through each level to create the subplot
    for i, level in enumerate(obesity_levels):
        subset = df[df['Obesity level'] == level]
        sns.scatterplot(data=subset, x='Age', y='Weight', ax=axes[i], hue='Obesity level', palette='Set2')
        axes[i].set_title(f'Obesity Level: {level}')
        axes[i].set_xlabel('Age')
        axes[i].set_ylabel('Weight')
        axes[i].grid()

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
scatterplot_obesity_level(df)
```

```python
# Plotting the histogram for Categorical graph, for the columns Transportation and Obesity level.

def histogram_obesity(df):
    '''
    Creates subplots for the mode of transport for each obesity level.
    '''
    obesity_levels = df['Obesity level'].unique()
    num_levels = len(obesity_levels)
    num_cols = (num_levels + 1) // 2 
    fig, axes = plt.subplots(nrows=2, ncols=num_cols, figsize=(5 * num_cols, 10), sharey=True)
    axes = axes.flatten()
    
    # Iterate through each to create the subplot
    
    for i, level in enumerate(obesity_levels):
        subset = df[df['Obesity level'] == level]
        axes[i].hist(subset['Transportation'], bins=5, alpha=0.7, edgecolor='black')
        axes[i].set_title(f'Obesity Level: {level}')
        axes[i].set_xlabel('Mode of Transport')
        axes[i].set_ylabel('Count')
        axes[i].grid()
        
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
histogram_obesity(df)
```

```python
# Plotting the boxplot

def boxplot_obesity(df):
    '''
    Creates an enhanced boxplot to visualize the relationship between high caloric food consumption and physical activity.
    '''
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df, x='Frequent consumption of high caloric food', y='Frequency of physical activity') 
    plt.title('Relationship Between High Caloric Food Consumption and Physical Activity', fontsize=16)
    plt.xlabel('Frequency of High Caloric Food Consumption', fontsize=14)
    plt.ylabel('Frequency of Physical Activity', fontsize=14)
    plt.xticks(rotation=45)
    plt.yticks(fontsize=12)  
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout() 
    plt.show()
```

```python
boxplot_obesity(df)
```

```python
# Finding the correlation of the dataset

df_corr = df1.corr(numeric_only = True)
df_corr
```

```python
# Plotting the Correlation Matrix Heatmap for the dataset

def plot_corr(corr):
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    heatmap = sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8}, linewidths=.5)
    plt.title('Correlation Matrix Heatmap', fontsize=16)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
plot_corr(df_corr)
```
